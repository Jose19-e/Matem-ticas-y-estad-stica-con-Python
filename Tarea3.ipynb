```python
#Comprobar grÃ¡ficamente si las siguientes ecuaciones tienen soluciÃ³n.
print("""
3x + 2y = 6
6x + 4y = âˆ’9 """)

x = np.linspace(-10, 10)
y1 = (6 - 3*x) / 2
y2 = (-9 - 6*x) / 4

plt.plot(x, y1, label='3x + 2y = 6')
plt.plot(x, y2, label='6x + 4y = -9')
plt.title('GrÃ¡fica de las ecuaciones Â¡SIN SOLUCIÃ“N!')
plt.grid(True)
plt.legend()
plt.show()
```

```python
x = np.linspace(-10, 10)
y1 = 3 - x
y2 = (1 - x) / 4

plt.plot(x, y1, label='x + y = 3')
plt.plot(x, y2, label='x - 4y = 1')
plt.title('GrÃ¡fica del sistema de ecuaciones Â¡CON SOLUCION!')
plt.grid(True)
plt.legend()
plt.show()
```

```python
A = np.array([[2, -1, 2], [2, 3, -1], [1, 2, 2]])
b = np.array([3, 0, -12])

solution = np.linalg.solve(A, b)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(solution[0], solution[1], solution[2], color='red', s=50, label='Punto de IntersecciÃ³n')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Punto de intersecciÃ³n del sistema de ecuaciones,Â¡CON SOLUCIÃ“N!')
ax.legend()

plt.show()

```
